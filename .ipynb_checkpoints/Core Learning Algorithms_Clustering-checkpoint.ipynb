#### <font color='red'>Clustering (叢) is a Machine Learning technique </font> that involves the grouping of data points. 
- data points that are in the same group should have similar properties and/or features
- data points in different groups should have highly dissimilar properties and/or features

#### Clustering 跟 Linear Regression，Classification 不同的地方在於，<font color='red'>cluster 沒有 label</font>。他只有很多很多的 data。  
#### 而目標就是將這些 data 分為 K 個 clusters。並預測新的資料點屬於哪一個 cluster。

#### Basic Algorithm for K-Means:
- STEP 1: 首先決定要設幾個 centroids(假設 K = 3)。
- STEP 2: 隨機把這三個 centroids 放置在圖形。
- STEP 3: 對於每個資料點，計算他們到三個 centroids 的距離，並再資料點上標示距離最短的那個 centroid 的編號。
- STEP 4: 之後，把屬於同一個 centroid 的所有資料點位置進行平均，把 centroid 放在那裡。
- STEP 5: 這時候 centroid 的位置可能跟一開始我們隨機放置的位置不同，因此，根據最新位置，重新調整資料點屬於哪個 centroid。
- STEP 6: 重複最後兩個步驟，直到沒有任何資料點被修改。 [<font color='red'>此時的 centroid 點就叫做 cluster</font>]
