#### 區分 AI，ML，以及 NN:
##### AI (Artificial Intelligence): 範圍最大。只要是人類 "規定" 演算法，並讓電腦執行，都稱做是運用 AI 的技術。
##### ML (Machine Learning): 範圍次大。現在人類不定規則，而是藉由一大筆資料讓電腦自己學習。
##### NN (Neural Network): 範圍最小。運用 ML，資料丟入後，只經過一層電腦自己學習的 Network，就產生結果。但 **NN 有很多層**。
---
#### 不管是 AI，ML，還是 NN，都是丟入已知的數據，其中數據的 <font color = 'red'>input 叫做 feature(特徵值)，ouput 叫做 label。</font>


### 三種 ML 方式:
- Supervised Learning
- Unsupervised Learning
- Reinforcement Learning

#### Supervised Learning: 就是將資料分為兩部分，其中一部分(包含 feature, label) 給電腦學習，另一部分拿來考電腦(也就是只給他 feature)。  
---
#### Unsupervised Learning: 我們就給電腦 feature，讓他嘗試自己把不同 feature 分類，這個過程稱為 <font color='red'>cluster</font>
---
#### Reinforcement Learning: 沒有任何資料，只有 <font color='red'> agent，environment，與 reward</font>
舉例

#### 上面的人叫 agent，他的目標是走到那面旗子。整個 agent 跟最終的目標(旗子)稱為 environment。電腦開始隨機的走，當他朝 flag 的方  
#### 像走，我們就給他 <font color='red'> +1 reward </font>
#### 相對的，如果他朝相反方向走，就給他 <font color='red'> -2 reward </font>。
#### 電腦的目標是最大化 reward，因此給他學習一段時間，他便能走到 flag。
