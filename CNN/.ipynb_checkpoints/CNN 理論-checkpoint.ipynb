## Convolutional Layer

Convolutional Layer VS Densly connected Layer:
目標是一樣的: 找到圖片裡的 pattern ，以用來辨認圖片。
但看的方式不一樣:
- Densly connected layer: 由於每個 neuron 的 data，來自前面的所有 neurons，因此一個 neuron 接著一個 neuron:  
   - 他 <font color='red'> 只能用 global 的方式看 </font>整張圖片。
   - 當我把照片翻個 180 度，電腦就無法辨認照片了。
- Convolutional layer 因為 layer 間不是 densly connected，因此:
   - 可以 <font color='red'> 用 local patterns 的方式看 </font>圖片。

![alt text](https://img.webmd.com/dtmcms/live/webmd/consumer_assets/site_images/article_thumbnails/reference_guide/cat_weight_ref_guide/1800x1200_cat_weight_ref_guide.jpg)

這是一張貓的圖片。神經網路的任務是辨認這張圖片是否是一隻貓。  
Densly connected layer 會將圖片的所有 pixels 看過一遍。  
Convolutional layer 則是將重點找出來，比如說:

以下一圖狗的例子來說，如果用 densly connected layer，他只會認定在圖片左方偏上有一個 feature(即狗眼)。  
一但將照片轉向(如下二圖)，原本左方偏上的位置沒有狗眼的 feature，他就認定此圖不是狗。(但顯然圖片仍是隻狗)

我們會使用多個 convolutional layers 來 train model。  
舉例來說，如果辨認狗的圖片有三層 layers，那:
- 第一層可能負責找出 線條跟 edges。並把資料丟入第二層。
- 第二層 layer 再將這些線條轉為各種 shape 或多邊形。並把資料丟入第三層。
- 第三層 layer 將這些 shapes 統合為一張圖片。

## Feature map:
A 3D tensor with: 
- 兩個特別的軸(axes) (width 和 height)。
- 一個 depth 軸。

Our convolutional layers take feature maps as their input and return a new feature map that reprsents the prescence(存在) of spcific filters from the previous feature map. (又叫做 response maps)  
<font color='red'> 也就是說數字代表的意思是某一個 feature 存在的可能。數字越大，這個 pixel 有此 feature 的機率就越高。</font>

## Filters:

> A filter is a m x n pattern of pixels that we are looking for in an image. The number of filters in a convolutional layer reprsents how many patterns each layer is looking for and what the depth of our response map will be. If we are looking for 32 different patterns/filters than our output feature map (aka the response map) will have a depth of 32. Each one of the 32 layers of depth will be a matrix of some size containing values indicating if the filter was present at that location or not.

我的理解:
    一個 convolutional layer 的 input 是一個 feature map。  
    傳進去後，我們要決定 filter 的數量與型態。(一個 filter 就是代表一個特別的 pattern，eg 狗的眼睛)。  
    而如果有 32 個 filters(也就是我們關心 32 個特別的 patterns)，則 feature map 的 depth = 32。  
    我們將 feature map 與 filter 做 dot product(內積)，得到另一個 feature map(又稱之為 response map)。  
    這個 response map 即為這層 convolutional layer 的 output。

## Padding(填充)

再用 filters 的時候，不能保證 feature map 的每一個 pixel 都會是 filter 的正中央。  
這代表 <font color='red'>response map(也就是跟 filter 做完內積的 feature map) 的大小一定不會跟原本的 feature map 一樣。</font>  
如果要解決這個問題，我們可以用 padding。  
其實就是將原來的 feature map 多增加邊界，讓所有 pixel 都可以當作正中央。  
<font color='red'>最終結果是 input feature map 與 output response map 的維度是一樣的。</font>

## Pooling(池化)

在訓練時會產生很多的 layers，有時資訊會太多。
如果要<font color='red'>降低資料的維度</font>，就要用 pooling。
Pooling 就是將 map 從左上角開始，每一次看 2*2 的資料，並根據 Pooling 的種類計算，丟進新的 map。   
** <font color='red'>通常 pooling 都是用 2乘以2 且 strides = 2 </font>。**
<font color='red'>如此一來就可以降低維度。</font>    
常見的 pooling 有:
- Max pooling: 取 2*2 map 裡的最大值。
- Min pooling: 取 2*2 map 裡的最小值。
- Average pooling: 取 2*2 map 裡的平均值。
