#### 一個 Neural Network 包含:
- Input layer
- Hidden layer
- Output layer

## <font color='red'>Input layer </font>:

#### 一個 Input layer 有許多個 neurons(神經元)，有多多呢? 那要看你要丟入神經網路的資料有多大。
----
#### 比如你將一張 28*28 pixels 的圖片都進去，那你需要 28*28=784 個 neurals(一個 neuron 配一個 pixel)

## <font color='red'>Output layer </font>:
#### Output layer 的 neurons 取決於你主觀認定。
----
#### 比如你在做 classifications，共有 5 個類別，那你認為是第一類的機率是 0.99，第二類的機率是 0.0005，第三類是...  
#### 總之，neurons 所代表的總和必須是 1(機率的和是 1)。
#### 這是 <font color='green'> 機率分配的概念</font>。

圖示

## <font color='red'>Hidden layer </font>

#### 在 input layer 與 output layer 間，就是 hidden layer。
#### 而 neuron 跟 neuron 間可以相連接。[用 <font color='green'>用 weight 實現 </font>]
#### 如果 layer 跟 layer 間的所有 neuron 都相連，稱為 densly connected layer。
#### 如果所有 layer 都是 densly connected layer，則整個 neural network 稱作 densly connected neural network。
---
#### 剛才提到的 weights 是可以修改的，他們是 trainable parameters。

##  <font color='red'> Bias </font>
#### Input layer 與 hidden layer 都會有一個 bias，而 bias 有一些特點:
- Bias 的 weight 都是 1。(也就是他跟別人連結時的 weight 都是 1)
- Bias 是 trainable parameter。
- Input layer 的 Bias 並不輸入資料。

Bias 圖示

#### <font color='red'> red 代表 0，blue 代表 1 </font>。
##### 我們就用圖示方法那樣加總。

## <font color='red'>Activation function</font>

#### 我們剛才這樣把資料乘上 sum，又加 bias，又進入另一層 layer，我們無法確保最終到達 output layer 的值會介於 0 到 1。  
#### 要知道，一開始時，<font color='green'> weights 跟 bias 都是 <font color='red'>random 的</font></font>。
---
#### 每個 neuron 都可以套用 activation function(因為每個 neuron 都是從別的 neurons 計算得來的)。
#### 那如果我最終的 outut 必須要在 0 到 1 之間，則 output layer 的那個 neuron，就可以套用 <font color='red'>Sigmoid </font>function。
---
#### 套用 activation function 的另一個好處是: <font color='red'> 增加維度 </font>。
##### 可以想像原本只有長寬的資料，如果多增加一個維度，那我就可以知道體積。

## <font color='red'>Loss function</font>:

#### 基本上，Machine Learning 在做的，就是把一堆資料丟進去模型，讓模型跑出 output，而我們自己會有正確答案。我們比較模型給的答案與正確答案的差距，並修改模型讓差距越來越小。
---
#### 問題時，怎麼知道那個差距是多少呢? 答案就是看 loss function 的值。
##### Loss function 的值越大，代表差距很大。

## <font color='red'>Gradient Descent 與 Backpropagation<font>:

![alt text](https://cdn-images-1.medium.com/max/1000/1*iU1QCnSTKrDjIPjSAENLuQ.png)

#### 所謂的 gradient descent，就是一個演算法，這個演送法的目的是找到 map 的 global minimum(換言之，loss function 的最小值)
#### gradient: 往 global minimum 的 <font color='red'> 那個方向 </font>。
#### 而 Backpropagation 就是修改 weights, bias 的過程(也是一個演算法) [叫 "back"propagation 嘛，所以是往後修改之前的 layer(hidden, input layer)]

## <font color='red'>Optimizer [較進階] </font>
#### 也是一種演算法。
You may sometimes see the term optimizer or optimization function. This is simply the function that implements the backpropagation algorithm described above. Here's a list of a few common ones.
- Gradient Descent
- Stochastic Gradient Descent
- Mini-Batch Gradient Descent
- Momentum
- Nesterov Accelerated Gradient
